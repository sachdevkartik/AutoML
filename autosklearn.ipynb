```python
y = df["class"]
df = df.drop(['class'], axis=1)
df.head()
```

```python
import numpy as np
df.replace([np.inf, -np.inf], np.nan, inplace=True)
print(f"{df.isna().any(axis=1).sum()} invalid rows dropped")
df.dropna(inplace=True)

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
myEncoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
columns_to_encode = ['V25',
                 'V83',
                 'V56',
                 'V52',
                 'V69',
                 'V47',
                 'V26',
                 'V99',
                 'V100',
                 'V74']
myEncoder.fit(df[columns_to_encode])

df = pd.concat([df.drop(columns_to_encode, 1),
          pd.DataFrame(myEncoder.transform(df[columns_to_encode]))], axis=1).reindex()
df.head()
```

```python
import pandas as pd
from sklearn import preprocessing

x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
df.head()
```

```python
train_df = df
np.unique(y)
```

```python
from sklearn import preprocessing
import numpy as np
le = preprocessing.LabelEncoder()
le.fit(np.unique(y))
trans_y = le.transform(y)
# inv_y = le.inverse_transform(trans_y)
# np.unique(inv_y)
# for x in range(len(y)):
#     print (y[x])
X = train_df
np.unique(y)
```

```python
import sklearn.datasets
import sklearn.metrics
import sklearn.model_selection
from sklearn.model_selection import train_test_split

def get_data_and_scoring_function(X, y):
    
    X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(
      X, trans_y, random_state=42, stratify=y
    )
    print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
    
    def scoring_function(estimator):        
        predictions = estimator.predict_proba(X_test)[:, 1]
        print(predictions)
        return sklearn.metrics.f1_score(y_test, predictions)
    def train_scoring_function(estimator):        
        predictions = estimator.predict_proba(X_train)[:, 1]
        print(predictions)
        return sklearn.metrics.f1_score(y_train, predictions)
    return X_train, y_train, scoring_function, train_scoring_function, X_test, y_test

X_train, y_train, scoring_function, train_scoring_function, X_test, y_test = get_data_and_scoring_function(X, y)
```
